```python
from pinecone import Pinecone, ServerlessSpec
import os # Nützlich, um den API Key sicher aus Umgebungsvariablen zu laden

# --- Konfiguration ---
INDEX_NAME = "firmenwissen-gcp"
VECTOR_DIMENSION = 1536  # Beispiel: Dimension für OpenAI text-embedding-ada-002

# API Key sicher laden
# Ersetzen Sie "YOUR_API_KEY" durch Ihren tatsächlichen Schlüssel 
# oder besser: os.environ.get("PINECONE_API_KEY")
API_KEY = "YOUR_API_KEY" 

# --- Index-Spezifikation (GCP Serverless) ---
gcp_serverless_spec = ServerlessSpec(
    cloud="gcp",
    region="us-west1-gcp"  # Passen Sie die Region an Ihre GCP-Dienste an!
)

# --- Index erstellen ---
try:
    # 1. Pinecone Client initialisieren
    pc = Pinecone(api_key=API_KEY)
    
    # 2. Prüfen, ob der Index bereits existiert
    if INDEX_NAME not in pc.list_indexes().names:
        
        # 3. Index erstellen
        print(f"Erstelle neuen Index: {INDEX_NAME}...")
        pc.create_index(
            name=INDEX_NAME,
            dimension=VECTOR_DIMENSION,
            metric='cosine',
            spec=gcp_serverless_spec
        )
        print(f"Index '{INDEX_NAME}' erfolgreich erstellt.")
    else:
        print(f"Index '{INDEX_NAME}' existiert bereits. Überspringe Erstellung.")

except Exception as e:
    print(f"Fehler bei der Index-Erstellung: {e}")
```
